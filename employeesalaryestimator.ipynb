```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create sample dataset
data = {
    'Education_Level': ['Bachelors', 'Masters', 'PhD', 'Bachelors', 'Masters', 'PhD', 'Bachelors', 'Masters'],
    'Years_of_Experience': [3, 7, 10, 2, 8, 12, 5, 6],
    'Job_Title': ['Data Analyst', 'Data Scientist', 'Researcher', 'HR Manager', 'Data Scientist', 'Researcher', 'HR Manager', 'Data Analyst'],
    'Location': ['New York', 'San Francisco', 'Boston', 'Austin', 'San Francisco', 'Boston', 'Austin', 'New York'],
    'Salary': [70000, 120000, 135000, 65000, 125000, 140000, 68000, 90000]
}

df = pd.DataFrame(data)
df.head()
```

```python
df['Education_Level'] = df['Education_Level'].astype('category').cat.codes
df['Job_Title'] = df['Job_Title'].astype('category').cat.codes
df['Location'] = df['Location'].astype('category').cat.codes
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Education_Level', y='Salary', data=df)
plt.title('Salary vs Education Level')
plt.xlabel('Education Level (0=Bachelors, 1=Masters, 2=PhD)')
plt.ylabel('Salary ($)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Define features and target variable
X = df[['Education_Level', 'Years_of_Experience', 'Job_Title', 'Location']]
y = df['Salary']

# Initialize model
estimator = LinearRegression()
estimator.fit(X, y)
```

```python
# Sample input: Masters, 5 years experience, Data Scientist, located in San Francisco
input_profile = np.array([[1, 5, 1, 1]])  # Masters, 5 yrs, Data Scientist, San Francisco
estimated_salary = estimator.predict(input_profile)

print(f"Estimated Compensation: ${estimated_salary[0]:.2f}")
```
